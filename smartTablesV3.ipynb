```python
#Module
import pandas as pd #Verarbeiten der Daten
from datetime import datetime #aktuelles Datum ausgeben
import json #Dictionary der config als .json ausgeben
import copy #deepcopy für nested dictionarys im Config Element

#File Dialog
import tkinter as tk
from tkinter import filedialog
```


# Daten laden

# Funktionen und Templates

```python
#TEMPLATES für neue "CONFIG"-Elemente der SmartTables als Dictionary
#GUID wird grundsätzlich als Spalte mit angegeben!

configTemplate = {
    "version":3,
    "index":1,
    "grid":{},
    "pivot":{},
    "gridLive":{
        "columns":[{
            "disableHtmlEncode":True,
            "allowSorting":True,
            "allowResizing":True,
            "allowFiltering":True,
            "allowGrouping":True,
            "allowReordering":True,
            "showColumnMenu":True,
            "enableGroupByFormat":False,
            "allowEditing":True,
            "showInColumnChooser":True,
            "allowSearching":True,
            "autoFit":False,
            "sortDirection":"Descending",
            "index":"0","field":"0",
            "propertyName":"GUID",
            "propertySetName":"BIMcollab",
            "type":"string",
            "visible":False,
            "minWidth":140,
            "foreignKeyField":"0",
            "headerText":"GUID"
        },{
            "disableHtmlEncode": True,
            "allowSorting": True,
            "allowResizing": True,
            "allowFiltering": True,
            "allowGrouping": True,
            "allowReordering": True,
            "showColumnMenu": True,
            "enableGroupByFormat": False,
            "allowEditing": True,
            "showInColumnChooser": True,
            "allowSearching": True,
            "autoFit": False,
            "sortDirection": "Descending",
            "index": "1",
            "field": "1",
            "propertyName": "GUID",
            "propertySetName": "Summary",
            "type": "string",
            "visible": True,
            "minWidth": 140,
            "foreignKeyField": "1",
            "headerText": "GUID"
        }],
        "filterSettings":{"type":"Menu","columns":[]},
        "sortSettings":{"columns":[]},
        "aggregationSettings":{"merge":False,"columns":[
            {
                "field": "1",
                "propertyName": "GUID",
                "propertySetName": "Summary",
                "aggregation": "None"
            }
        ]}
    },
    "pivotLive":{},
    "filter":"SmartViewHex",
    "selected_properties":[{"label": "GUID","propertySet": "Summary"}]
}

#config.gridlive.column Template
gridLiveColumnTemplate = {
    "disableHtmlEncode":True,
    "allowSorting":True,
    "allowResizing":True,
    "allowFiltering":True,
    "allowGrouping":True,
    "allowReordering":True,
    "showColumnMenu":True,
    "enableGroupByFormat":False,
    "allowEditing":True,
    "showInColumnChooser":True,
    "allowSearching":True,
    "autoFit":False,
    "sortDirection":
    "Descending",
    "index":"0",
    "field":"0",
    "propertyName":"GUID",
    "propertySetName":"Summary",
    "type":"string",
    "visible":True,
    "minWidth":140,
    "foreignKeyField":"0",
    "headerText":"GUID"
}

#config....aggregationSetting Template
aggregationSettingTemplate = {
    "field":"1",
    "propertyName":"GUID",
    "propertySetName":"Summary",
    "aggregation":"None"
}


#config.selected_Properties Template
selectedPropertyTemplate = {
    "label":"GUID",                        
    "propertySet":"Summary"
}
```

```python
#Funktion zum erstellen der eingebetteten SmartView
#Wiedergabe als hex encoded string

def newFilterAsHex(element, predefinedType):

    sview = """<SMARTVIEW>
    <TITLE>Filter components</TITLE>
    <DESCRIPTION></DESCRIPTION>
    <CREATOR></CREATOR>
    <CREATIONDATE>2022-07-30T14:29:04</CREATIONDATE>
    <MODIFIER></MODIFIER>
    <MODIFICATIONDATE>2022-07-30T14:29:04</MODIFICATIONDATE>
    <GUID></GUID>
    <RULES>
        <RULE>
            <IFCTYPE>Any</IFCTYPE>
            <PROPERTY>
                <NAME>IFC Element</NAME>
                <PROPERTYSETNAME>Summary</PROPERTYSETNAME>
                <TYPE>Summary</TYPE>
                <VALUETYPE>StringValue</VALUETYPE>
                <UNIT>None</UNIT>
            </PROPERTY>
            <CONDITION>
                <TYPE>Is</TYPE>
                <VALUE>{ifcElementName}</VALUE>
            </CONDITION>
            <ACTION>
                <TYPE>And...</TYPE>
            </ACTION>
        </RULE>
        <RULE>
            <IFCTYPE>Any</IFCTYPE>
            <PROPERTY>
                <NAME>Predefined Type</NAME>
                <PROPERTYSETNAME>Summary</PROPERTYSETNAME>
                <TYPE>Summary</TYPE>
                <VALUETYPE>StringValue</VALUETYPE>
                <UNIT>None</UNIT>
            </PROPERTY>
            <CONDITION>
                <TYPE>Is</TYPE>
                <VALUE>{predefinedType}</VALUE>
            </CONDITION>
            <ACTION>
                <TYPE>Add</TYPE>
            </ACTION>
        </RULE>
    </RULES>
    <INFORMATIONTAKEOFF>
        <PROPERTYSETNAME>None</PROPERTYSETNAME>
        <PROPERTYNAME>None</PROPERTYNAME>
        <OPERATION>0</OPERATION>
    </INFORMATIONTAKEOFF>
</SMARTVIEW>\u000A""".format(ifcElementName=element, predefinedType=predefinedType)
    
    return sview.encode("utf-8").hex().upper()
```

```python
#Funktion zur Erstellung des CONFIG elements aus den Templates

def newTableConfig(element, predefinedType, tableIndex, attributes):

    #deepcopy necessary for nested dict
    config = copy.deepcopy(configTemplate)
    config["index"] = tableIndex
    config["filter"] = newFilterAsHex(element, predefinedType)
    
    for index, row in attributes.iterrows():
        
        #Index angepasst (1 based) und um 1 erhöht wegen GUID im Template
        index+=2
        pset = row.Pset
        attribute = row.Attribut
        
        #copy template dictionaries and fill out values
        gridLiveColumn = gridLiveColumnTemplate.copy()
        gridLiveColumn["propertyName"] = attribute
        gridLiveColumn["propertySetName"] = pset
        gridLiveColumn["index"] = str(index)
        gridLiveColumn["field"] = str(index)
        gridLiveColumn["foreignKeyField"] = str(index)
        gridLiveColumn["headerText"] = attribute
               
        aggregationSetting = aggregationSettingTemplate.copy()
        aggregationSetting["field"] = str(index)
        aggregationSetting["propertyName"] = attribute
        aggregationSetting["propertySetName"] = pset      
        
        selectedProperty = selectedPropertyTemplate.copy()
        selectedProperty["label"] = attribute
        selectedProperty["propertySet"] = pset
        
        #add dicts to config
        config["gridLive"]["columns"].append(gridLiveColumn)
        config["gridLive"]["aggregationSettings"]["columns"].append(aggregationSetting)
        config["selected_properties"].append(selectedProperty)
        

    return json.dumps(config)
```

# smartTables.bcl erstellen

```python
#Datei schreiben und speichern

date = datetime.now().strftime("%d.%m.%Y")
creationDate = datetime.now().strftime("%Y-%m-%dT%H:%M:%S")

with open("SmartTables.bcl", "w", encoding="utf-8") as xmlFile:

    #File header
    xmlFile.write("""<?xml version=\"1.0\"?>
<bimcollabtablesfile>
    <version>3</version>
    <applicationversion>Win - Version: 6.0 (build 6.0.0.4)</applicationversion>
</bimcollabtablesfile>

<TABLESETS>""")
    
    #Ein Tableset pro Gewerk
    for gewerk in df.gewerk.unique():
        tableSetTitle = "Attribute - {}".format(gewerk)
        xmlFile.write("""
    <TABLESET>
        <TITLE>{}</TITLE>
        <DESCRIPTION>Modifiziert: Diese Listen zeigen die geforderten Attribute gemäß Parameterliste vom {}.</DESCRIPTION>
        <TABLES>""".format(tableSetTitle, date))

        #Eine Tabelle je ifcType und Predefined Type
        for tableIndex, row in df[df.gewerk == gewerk].drop_duplicates(["element", "predefinedType"]).reset_index().iterrows():          
            tableTitle = "{} - {} - {}".format(row.element, row.beschreibung, row.typ)
            description = row.beschreibung
            attributes = df[(df.gewerk==gewerk) & (df.element==row.element) & (df.predefinedType==row.predefinedType)][["Pset", "Attribut"]]
            newConfig = newTableConfig(row.element, row.predefinedType, tableIndex, attributes)

            xmlFile.write("""
            <TABLE>
                <TITLE>{}</TITLE>
                <DESCRIPTION>{}</DESCRIPTION>
                <CREATOR>Formitas AG</CREATOR>
                <CREATIONDATE>{}</CREATIONDATE>
                <TABLETYPE>GridView</TABLETYPE>
                <CONFIG>{}</CONFIG>
            </TABLE>""".format(tableTitle ,description, creationDate, newConfig))

        xmlFile.write("""
            </TABLES>
        </TABLESET>
    """)     
    xmlFile.write("</TABLESETS>")
```
